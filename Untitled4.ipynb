```python
import kagglehub
import pandas as pd
import os

# Download dataset from Kaggle
path = kagglehub.dataset_download('teertha/ushealthinsurancedataset')
print("Dataset path:", path)

# List downloaded files
for f in os.listdir(path):
    print(" →", f)

# Load CSV
csv_file = os.path.join(path, 'insurance.csv')
df = pd.read_csv(csv_file)

# Ensure numeric types
df['charges']  = pd.to_numeric(df['charges'],  errors='coerce')
df['age']      = pd.to_numeric(df['age'],      errors='coerce')
df['bmi']      = pd.to_numeric(df['bmi'],      errors='coerce')
df['children'] = pd.to_numeric(df['children'], errors='coerce')
df.dropna(inplace=True)

print(f"\n✅ Dataset loaded: {df.shape[0]} rows × {df.shape[1]} columns")
print(df.dtypes)
df.head()
```

```python
df_proc = df.copy()

# Binary encoding
df_proc['smoker'] = (df_proc['smoker'] == 'yes').astype(int)  # yes→1, no→0
df_proc['sex']    = (df_proc['sex'] == 'male').astype(int)    # male→1, female→0

# One-hot encode region (drop first to avoid multicollinearity)
df_proc = pd.get_dummies(df_proc, columns=['region'], drop_first=True)

# Log-transform target
df_proc['log_charges'] = np.log1p(df_proc['charges'])
df_proc.drop(columns=['charges'], inplace=True)

# Features and target
FEATURE_COLS = [c for c in df_proc.columns if c != 'log_charges']
X = df_proc[FEATURE_COLS]
y = df_proc['log_charges']

# Train-test split (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# Standard scaling
scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print(f"Train : {X_train_sc.shape}")
print(f"Test  : {X_test_sc.shape}")
print(f"Features: {FEATURE_COLS}")
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression' : Ridge(alpha=1.0),
    'Random Forest'    : RandomForestRegressor(
                            n_estimators=200, random_state=42, n_jobs=-1)
}

print("── 5-Fold Cross-Validation R² Scores ──")
for name, model in models.items():
    scores = cross_val_score(model, X_train_sc, y_train,
                             cv=kfold, scoring='r2')
    print(f"{name:22s} | Mean R² = {scores.mean():.4f} ± {scores.std():.4f}")

# Train on full training set
for name, model in models.items():
    model.fit(X_train_sc, y_train)

print("\n✅ All 3 models trained!")
```

```python
def evaluate(name, model, X, y):
    y_pred_log = model.predict(X)
    y_pred     = np.expm1(y_pred_log)   # reverse log-transform
    y_true     = np.expm1(y)
    return {
        'Model'   : name,
        'MAE ($)' : round(mean_absolute_error(y_true, y_pred), 2),
        'RMSE ($)': round(np.sqrt(mean_squared_error(y_true, y_pred)), 2),
        'R²'      : round(r2_score(y_true, y_pred), 4),
        'y_pred'  : y_pred,
        'y_true'  : y_true
    }

results = [evaluate(n, m, X_test_sc, y_test) for n, m in models.items()]

summary = pd.DataFrame([
    {k: v for k, v in r.items() if k not in ('y_pred','y_true')}
    for r in results
]).set_index('Model')

print("── Test Set Results (original USD scale) ──")
display(summary)
```

```python
# Actual vs Predicted
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, r in zip(axes, results):
    lims = [min(r['y_true'].min(), r['y_pred'].min()),
            max(r['y_true'].max(), r['y_pred'].max())]
    ax.scatter(r['y_true'], r['y_pred'],
               alpha=0.4, s=20, color='steelblue', edgecolors='none')
    ax.plot(lims, lims, 'r--', lw=1.5, label='Perfect fit')
    ax.set_title(f"{r['Model']}\nR² = {r['R²']}", fontweight='bold')
    ax.set_xlabel('Actual ($)')
    ax.set_ylabel('Predicted ($)')
    ax.xaxis.set_major_formatter(
        mticker.FuncFormatter(lambda x,_: f'${x/1000:.0f}k'))
    ax.yaxis.set_major_formatter(
        mticker.FuncFormatter(lambda x,_: f'${x/1000:.0f}k'))
plt.tight_layout()
plt.savefig('eval_actual_vs_predicted.png', bbox_inches='tight')
plt.show()

# Feature Importance
rf      = models['Random Forest']
feat_df = pd.DataFrame({
    'Feature'   : FEATURE_COLS,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=True)

colors_fi = ['#e74c3c' if f == 'smoker' else '#3498db'
             for f in feat_df['Feature']]

plt.figure(figsize=(9, 5))
plt.barh(feat_df['Feature'], feat_df['Importance'],
         color=colors_fi, edgecolor='white')
plt.title('Feature Importances — Random Forest  (Red = Smoker)',
          fontweight='bold')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('eval_feature_importance.png', bbox_inches='tight')
plt.show()
```
